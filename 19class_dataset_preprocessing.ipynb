```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import shuffle
import pickle
```

```python
path_dir = os.getcwd()+"/19class_dataset/19class/"
```

```python
CATEGORIES = os.listdir(path_dir)
```

```python
IMG_SIZE = 600
```

```python
data=[]
```

```python
for category in CATEGORIES:
    path = os.path.join(path_dir,category+"/")
    for img in os.listdir(path):
        try:
            img_arr = cv2.imread(os.path.join(path,img))
            img_resized  = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
            data.append([img_resized,CATEGORIES.index(category)])
        except Exception as e:
            pass
```

```python
len(data)
```

```python
shuffle(data)
```

```python
X = []
Y = []
```

```python
for features,labels in data:
    X.append(features)
    Y.append(labels)
```

```python
dataset = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,3)
```

```python
labelset = np.array(Y).reshape(-1,1)
```

```python
labelset[0][0]
```

```python
plt.imshow(dataset[0])
plt.title("Img from DATASET")
plt.show()
```

```python
x_training = open("19class_dataset.pickle","wb")
pickle.dump(dataset,x_training)
```

```python
y_labels = open("19class_labels.pickle","wb")
pickle.dump(labelset,y_labels)
pickle.close()
```

```python
y_labels = open("19class_labelset.pickle","wb")
pickle.dump(labelset,y_labels)
y_labels.close()
```
